```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import LabelEncoder
```

```python
# Fusion des données
merged_data = pd.merge(fact_trainings, dim_employees, left_on='FK_Employees', right_on='CODE', how='left')
merged_data = pd.merge(merged_data, dim_trainings_cleaned, left_on='Fk_Training', right_on='course_id', how='left')
```

```python
merged_data
```

```python
# Création de la matrice utilisateur-formation avec le nom de l'employé comme index
user_course_matrix = merged_data.pivot_table(index='CODE', columns='course_Name', aggfunc='size', fill_value=0)
```

```python
user_course_matrix
```

```python
# Calcul de la similarité cosinus entre les utilisateurs (employés)
user_similarity = cosine_similarity(user_course_matrix)
```

```python
# Création d'un dictionnaire de correspondance entre le CODE et le NAME_EMP
code_to_name = merged_data.set_index('CODE')['NAME_EMP'].to_dict()
```

```python
def recommend_courses(employee_code, top_n=5):
    if employee_code not in user_course_matrix.index:
        return f"L'employé avec le code '{employee_code}' n'a pas été trouvé dans la base de données."
    
    # Trouver l'indice de l'employé dans la matrice
    user_index = user_course_matrix.index.get_loc(employee_code)
    
    # Calculer les similarités avec tous les autres utilisateurs
    similarities = user_similarity[user_index]
    
    # Trier les indices des utilisateurs par similarité décroissante
    similar_users_indices = similarities.argsort()[::-1]
    
    # Récupérer les formations suivies par les utilisateurs similaires
    similar_users_courses = user_course_matrix.iloc[similar_users_indices[1:]]  # Exclure l'employé lui-même
    
    # Trouver les formations non suivies par l'employé cible mais suivies par les utilisateurs similaires
    courses_to_recommend = similar_users_courses.sum(axis=0)
    courses_to_recommend = courses_to_recommend[courses_to_recommend == 0].index  # Formations non suivies
    
    # Sélectionner les meilleures formations à recommander
    top_courses = courses_to_recommend[:top_n]
    
    return [(course, code_to_name[employee_code]) for course in top_courses] if not top_courses.empty else ["Aucune recommandation disponible pour cet employé."]
```
