```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
coursera = pd.read_csv("Coursera_trainings.csv")
```

```python
coursera.head()
```

```python
coursera = coursera.drop(columns=['id_course'])
```

```python
col = 'Difficulty Level'  
print(col)

plt.figure(figsize=(15, 4))

plt.subplot(1, 1, 1)
sns.countplot(x=coursera[col])
plt.xlabel(col)
plt.ylabel('Count')
plt.title(f'Distribution of {col}')
plt.xticks(rotation=45)  

plt.tight_layout()
plt.show()
```
