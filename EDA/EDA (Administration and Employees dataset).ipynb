```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
numerical_data = admin.select_dtypes(include=['int64', 'float64'])
```

```python
# Standardize the data 
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numerical_data)
```

```python
# Apply ACP
pca = PCA()
pca_result = pca.fit_transform(scaled_data)
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
```

```python
# Plot the explained variance
cumulative_variance_ratio = explained_variance_ratio.cumsum()
plt.plot(range(1, len(cumulative_variance_ratio) + 1), cumulative_variance_ratio, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance by Principal Components')
plt.show()
```

```python
# Apply PCA with the chosen number of components
chosen_components = 3  # Adjust based on the plot or your criteria
pca = PCA(n_components=chosen_components)
final_result = pca.fit_transform(scaled_data)
```

```python
# Create a DataFrame with the principal components
columns = [f'PC{i+1}' for i in range(chosen_components)]
pca_df = pd.DataFrame(final_result, columns=columns)
```

```python
# Concatenate the principal components DataFrame with the original DataFrame if needed
result_with_original_data = pd.concat([admin, pca_df], axis=1)
```

```python

# Create a DataFrame with variable loadings (coefficients)
loadings_df = pd.DataFrame(pca.components_.T, columns=columns, index=numerical_data.columns)
```

```python
# Plot the biplot
plt.figure(figsize=(10, 10))
```

```python
# Plot principal components on the x-y plane
for i in range(chosen_components):
    plt.arrow(0, 0, pca.components_[0, i], pca.components_[1, i], head_width=0.05, head_length=0.1, fc='red', ec='red')
    plt.text(pca.components_[0, i], pca.components_[1, i], columns[i], color='red')
```

```python
# Plot variable loadings (original features) on the biplot
for feature in numerical_data.columns:
    x = loadings_df.loc[feature, 'PC1']
    y = loadings_df.loc[feature, 'PC2']
    plt.arrow(0, 0, x, y, head_width=0.05, head_length=0.1, fc='blue', ec='blue')
    plt.text(x, y, feature, color='blue')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Biplot of PCA')

plt.show()
```
